```python
from bs4 import BeautifulSoup
import requests
from wordcloud import WordCloud
import pandas as pd
```

```python
def getTitle(soup):
  # Find the title
  try:
    result = soup.find("h1", class_="article-title")
    return result.text.strip()
  except:
    pass
```

```python
def getDate(soup):
  # Find the title
  try:
    result = soup.find("div", class_="article-date")
    return result.text.split()[0]
  except:
    pass
```

```python
def getLabel(soup):
  # Find the title
  try:
    result = soup.find("ol", class_="breadcrumb")
    return result.text.split("\n")[3]
  except:
    pass
```

```python
def getViewNumber(soup):
  # Find the title
  try:
    result = soup.find("span", class_="viewcount-badge")
    return result.text.split()[0]
  except:
    pass
```

```python
def getContent(soup):
  # Find the content
  try:
    text = ""
    result = soup.find("div", itemprop="articleBody")
    paragraphs = result.find_all("p")
    for paragraph in paragraphs:
      text = text + paragraph.text.strip() #+ "\n"
    return text
  except:
    pass
```

```python
def getList(soup):
  # Find the content
  try:
    text = soup
    result = soup.find("section", {"class":"mod_group-columns"})
    # paragraphs = result.find_all("div")
    # for paragraph in paragraphs:
    #   text = text + paragraph.text.strip() #+ "\n"
    return result
  except:
    pass
```

```python
def scraping(url):
  page = requests.get(url)
  soup = BeautifulSoup(page.content, "html.parser")
  date = getDate(soup)
  label = getLabel(soup)
  views = getViewNumber(soup)
  title = getTitle(soup)
  text = getContent(soup)
  return [url, title, text, date, views, label]
```

```python
url = "https://www.taiwannews.com.tw/en/news/"
startId = 4852699

import time
from tqdm.notebook import tqdm
id = startId
df = pd.DataFrame(columns =['url', 'title', 'text', 'date', 'views', 'label'])
for i in tqdm(range (1000)):
  time.sleep(1)
  rowData = scraping(url+str(id))
  df.loc[len(df.index)] = rowData
  id = id+1
  lastId = id
```

```python
fileName = str(startId) + "-" + str(lastId-1) + ".csv"
df.to_csv(fileName, index=False)
```
